```
import cv2
import numpy as np
from skimage.color import rgb2gray
from PIL import Image
import matplotlib.pyplot as plt
from google.colab.patches import cv2_imshow
img_gray = plt.imread("/content/frankfurt_000000_000294_gtFine_labelTrainIds.png")#输入标签文件
#img_gray = rgb2gray(img)#这里还是要先灰度化

img_binary = np.where(img_gray >= 0.01, 255, 0)
#cv2.imwrite('lena.png',img_binary，cmap="gray")
plt.imshow(img_binary,cmap="gray")# 展示二值图
import matplotlib
plt.imsave('plt_binary.png', img_binary) #plt格式下的二值图
plt.savefig('plt_x_y.png')
image = cv2.imread('plt_binary.png', cv2.IMREAD_GRAYSCALE)
cv2.imwrite('binary_imgsave.png', image)
# 图片二值化
from PIL import Image
img = Image.open('/content/frankfurt_000000_000294_gtFine_color.png')
src = cv2.imread("/content/frankfurt_000000_000294_gtFine_color.png")
cv2_imshow(src)
 
# 模式L”为灰色图像，它的每个像素用8个bit表示，0表示黑，255表示白，其他数字表示不同的灰度。
Img = img.convert('L')
Img.save("L_photo.jpg")
src = cv2.imread("L_photo.jpg")
cv2_imshow(src)
 
# 自定义灰度界限，大于这个值为黑色，小于这个值为白色
threshold = 24
 
table = []
for i in range(256):
  if i < threshold:
    table.append(0)
  else:
    table.append(1)
 
# 图片二值化
photo = Img.point(table, '1')
photo.save("modify.jpg")
src = cv2.imread("modefy.jpg")


```

```
import cv2
from google.colab.patches import cv2_imshow

#读取图片
src = cv2.imread('/content/frankfurt_000000_000294_gtFine_labelTrainIds.png')

#灰度图像处理
GrayImage = cv2.cvtColor(src,cv2.COLOR_BGR2GRAY)

#二进制阈值化处理
r, b = cv2.threshold(GrayImage, 0, 255, cv2.THRESH_BINARY)
print(r)
print(b)

# 保存图片:就是把数组保存成图片
import PIL

from PIL import Image
im = Image.fromarray(b)
im.save("test_result.png")

#显示图像
cv2_imshow(src)
cv2_imshow(b)


#等待显示
cv2.waitKey(0)
cv2.destroyAllWindows()
```
